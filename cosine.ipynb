````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
````

````python
x = np.linspace(0,2*np.pi,100_000)
y = np.cos(x)
````

````python
x_train, x_val, x_test = (x[:int(0.81*len(x))],  # 0.81 because 0.9^2
                          x[int(0.81*len(x)):int(0.9*len(x))], 
                          x[int(0.9*len(x)):])
````

````python
y_train, y_val, y_test = (y[:int(0.81*len(y))], 
                          y[int(0.81*len(y)):int(0.9*len(y))], 
                          y[int(0.9*len(y)):])
````

````python
plt.plot(x_train,y_train, label="Train")
plt.plot(x_val, y_val, label="Val")
plt.plot(x_test, y_test, label="Test")
plt.legend(loc='upper center');
plt.xlabel("Timesteps");
plt.ylabel("Value");
````
